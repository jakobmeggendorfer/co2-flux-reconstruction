```python
import numpy as np
from scipy.stats import pearsonr
import os

data_path = "../../../../../../../Volumes/T7 Shield/preprocessed_new_exp1"
feature_files = [os.path.join(data_path, f"{i}_features.npy") for i in range(1, 733)]
target_files = [os.path.join(data_path, f"{i}_targets.npy") for i in range(1, 733)]

monthly_features = np.stack([np.load(fp) for fp in feature_files])
monthly_targets = np.stack([np.load(fp) for fp in target_files])


n_features = monthly_features.shape[-1]
pearson_scores = []


# x_flat = monthly_features[:, :, :].flatten()
y_flat = monthly_targets[...,2].flatten()
x_flat = monthly_features.reshape(-1, n_features)
mask = x_flat[:,11] == 1

# Remove tmask (index 11) from x_flat
x_flat = np.delete(x_flat, 11, axis=1)
n_features = n_features - 1

y_flat = y_flat[mask]
x_flat = x_flat[mask]

for f in range(n_features):
    r, _ = pearsonr(x_flat[:,f], y_flat)
    pearson_scores.append(r)

# Convert to numpy array for convenience
pearson_scores = np.array(pearson_scores)

# Display results
for i, r in enumerate(pearson_scores):
    print(f"Feature {i}: Pearson r = {r:.4f}")
```

```python
import matplotlib.pyplot as plt

features = ['SST', 'SAL', 'ice_frac', 'mixed_layer_depth', 'heat_flux_down', 'water_flux_up', 'stress_X', 'stress_Y', 'currents_X', 'currents_Y',
            'month','year','nav_lat', 'nav_lon','global_co2']

sorted_indices = np.argsort(pearson_scores)
sorted_scores = np.array(pearson_scores)[sorted_indices]
sorted_names = np.array(features)[sorted_indices]

# Plot
plt.bar(range(len(sorted_scores)), sorted_scores)
plt.xlabel("Feature")
plt.ylabel("Global Pearson Correlation with CO Flux")
plt.title("Sorted Feature-wise Global Pearson Correlation")
plt.xticks(ticks=range(len(sorted_scores)), labels=sorted_names, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
